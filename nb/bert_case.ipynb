```python
import numpy as np
def convert_to_samples(value_span_lis, context, max_len):
    samples = []
    if not context:
        return samples
    for span in value_span_lis:
        begin = span[0]
        end = span[1]
        if begin >= end or end > len(context) or begin < 0:
            print("error span: %s->%s" % (begin, end))
            continue
        # trunction
        if end - begin > (max_len - 3) // 2 - 2:
            end = begin + (max_len - 3) // 2 - 2
        cut_context = context
        i=0
        if len(context) + (end - begin) > max_len - 3 - 2:
            # cut context
            i = begin
            j = end
            while (j - i) + (end - begin) < max_len - 3 - 2:
                if i > 0:
                    i -= 1
                if j < len(context):
                    j += 1
            cut_context = context[i:j]

        begin_offset = begin-(i-0)
        end_offset = end-(i-0)
            # 加点扰动
        left_disturbance = 0
        right_disturbance = 0
        if True:
            left_disturbance = np.random.choice([0, 1, 2], p=[0.5, 0.3, 0.2])
            right_disturbance = np.random.choice([0, 1, 2], p=[0.5, 0.3, 0.2])
        if begin_offset - left_disturbance >= 0:
            begin_offset = begin_offset - left_disturbance
        if end_offset + right_disturbance <= len(cut_context):
            end_offset = end_offset + right_disturbance
        
   

        text_a = cut_context[begin_offset:end_offset]
        text_b = cut_context[:begin_offset] + "【" + text_a + "】" + cut_context[end_offset:]

        tmp = {}
        tmp["text_a"] = text_a
        tmp["text_b"] = text_b
        tmp["span"] = (begin_offset, end_offset)
        samples.append(tmp)
    return samples
```
